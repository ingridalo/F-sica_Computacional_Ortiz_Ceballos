```julia
using SymPy
using PyPlot
```

```julia
#Definimos la función para el mapeo
function mapeoLog(x,m)
    return 1-(m)*(x^2)
    end;
```

```julia
#Definimos la derivada de la función anterior:
function derivMap(x,m)
    return 1-(2*m)*x
    end;
```

```julia
explyapunovArray=zeros(1000); #Este array contienene los exponentes de Lyapunov
mArray=linspace(0,5,1000); #Dado m dividimos el intervalo [0,5] en 1000 partes
x0=0.5; #Damos una condición inicial
```

```julia
n=10000 #Se toma una n muy grande porque en la ecuación n debe tender a infinito
x=x0

#Se hace un for para los exponentes de Lyapunov que guarda el exponente:
for j in 1:length(explyapunovArray)
    suma=0 #hacemos que la suma empiece en 0
    for i in 1:n-1 
        suma+=log(abs(derivMap(x,mArray[j])))
        x=mapeoLog(x,mArray[j])
    end
    explyapunovArray[j]=suma/n
end
```

```julia
#GRAFICAMOS los exponentes de Lyapunov en función de m
grid("on");
plot(mArray,explyapunovArray);
```
