### Método de Newton.

```julia
#Ejercicio 1  esta función usa el método de bisección para poder graficar cómo convergen.

function newton(init::Float64)
    list=zeros(Complex64,100); #seguí usando complex en caso de que marcara error en el dominio
    x=init;
    for i in 1:100
        f=(x^3-x^2-x-1)
        df=(3x^2x-1) 
        x=x-f/df;
        list[i]=x;
    end
    return list
end;
```

```julia
#con esto se enlistan las iteraciones
list1=newton(0.1)
list2=newton(0.9)
list3=newton(1.5)
list4=newton(3.1)
```

```julia
using PyPlot
```

```julia
#con este comando le decimos que grafique las listas, como no hay parte imaginaria, le quité el "real(list(n))" del ejemplo de la clase)

plot(list1)
plot(list2)
plot(list3)
plot(list4)
show()
```

```julia
#Graficamos el error, con list[100] se le está diciendo que nos dé la última iteración. Nuevamente omití el "real(list(n))" porque no hay números complejos.
plot(((list1-list1[100])./list1[100]))
plot(((list2-list2[100])/list2[100]))
plot(((list3-list3[100])./list3[100]))
plot(((list4-list4[100])./list4[100]))
show()
```

### Método de Bisección

```julia
function biseccion(a,b)
    list=zeros(Complex64,100); #seguí usando complex en caso de que marcara error en el dominio
       
    p = (a+b)/2
    f(x) = x^3-x^2-x-1
    
    for i in 1:100
        if  f(a)*f(p) > 0
            a = p
        end
        
        if f(b)*f(p) > 0
            b = p
        end
        p = (a+b)/2
    
        list[i]=p
    end
    return list
end;
```

```julia
list5=(biseccion(0.1,2))
list6=(biseccion(0.9,2.2))
list7=(biseccion(1.5,3.5))
list8=(biseccion(3.1,4))
```

```julia
using PyPlot
```

```julia
plot(list5)
plot(list6)
plot(list7)
plot(list8)
show()
```
